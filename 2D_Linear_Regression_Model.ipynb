```python
import random

x = [0] * 10

y = [0] * 10

for i in range(10):
    x[i] += i + random.randint(-2, 2)*0.5 + 1
    y[i] += 2*i + random.randint(-2, 2)*0.5 + 2

print(f"{'i':>2} | {'x':>6} | {'y':>6}")
print("-" * 30)
for i in range(10):
    print(f"{i:>2} | {x[i]:>6.2f} | {y[i]:>6.2f}")
```

```python
x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)

b1 = sum((x[i] - x_mean) * (y[i] - y_mean) for i in range(len(x))) / sum((x[i] - x_mean) ** 2 for i in range(len(x)))
b0 = y_mean - b1 * x_mean;

print(f"b1 = {b1}")
print(f"b0 = {b0}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue', label='Data Points')

y_pred = [b0 + b1 * xi for xi in x]

plt.plot(x, y_pred, color='red', label='Best Fit Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression (2D)')
plt.legend()
plt.grid(True)
plt.show()


```
